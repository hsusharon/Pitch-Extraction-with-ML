```python
import crepe
from scipy.io import wavfile
#from os.path import expanduser
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the wav file (mono)
samplingFrequency, signalData = wavfile.read('wave_dataset/mic_F01_sa1.wav')

plt.figure(figsize=(17,5))
plt.title('Waveform')
plt.plot(signalData)
plt.xlabel('Time')
plt.show()
# Plot the signal read from wav file
plt.figure(figsize=(17,5))
plt.title('Spectrogram')
plt.specgram(signalData,Fs=samplingFrequency)
plt.xlabel('Time')
plt.show()
```

```python
import os
from scipy.io.wavfile import write

newdir = 'new_dataset/'
for dirname, _, filenames in os.walk('wave_dataset/'):
    for filename in filenames:
        #print(os.path.join(dirname, filename))
        sr, audio = wavfile.read(os.path.join(dirname,filename))
        newData = []
        for i in range(audio.size):
            if audio[i]>50  or  audio[i]<-50:
                newData.append(audio[i])
        newData = np.array(newData)
        new_filename = newdir + filename;
        print(new_filename, ' Create Succuessfully ')
        write(new_filename, sr, newData)
```

```python
samplingFrequency, signalData = wavfile.read('new_dataset/mic_F01_sa1.wav')

plt.figure(figsize=(17,5))
plt.title('Waveform')
plt.plot(signalData)
plt.xlabel('Time')
plt.show()
# Plot the signal read from wav file
plt.figure(figsize=(17,5))
plt.title('Spectrogram')
plt.specgram(signalData,Fs=samplingFrequency)
plt.xlabel('Time')
plt.show()
```

```python
import os

cvsdir = 'txtfile/'
for dirname, _, filenames in os.walk('new_dataset/'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        sr, audio = wavfile.read(os.path.join(dirname,filename))
        time, frequency, confidence, activation = crepe.predict(audio, sr, viterbi=True)
        a = np.column_stack((time, frequency, confidence))
        csv_file = cvsdir + filename +  '.csv'
        np.savetxt(csv_file, a,
                  ['%.3f', '%d', '%.6f'],
                  header='time,frequency,confidence',delimiter=',')
```
