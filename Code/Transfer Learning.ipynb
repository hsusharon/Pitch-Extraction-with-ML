```python
import crepe
from scipy.io import wavfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def zeropadding(array, SamplePerWindow):
    if(array.size%SamplePerWindow != 0):
        left = array.size%SamplePerWindow
        add = np.zeros(SamplePerWindow - left)
        new_arr = np.append(array, add)
        return new_arr
    else:
        return array
```

```python
import os

X = []
SamplePerWindow = 480 ## sample rate 48000/s 0.01sec/window
for dirname, _, filenames in os.walk('new_dataset/'):
    for filename in filenames:
        sr, audio = wavfile.read(os.path.join(dirname,filename))
        new_audio = zeropadding(audio, SamplePerWindow)
        print(os.path.join(dirname, filename), 'audio size: ', new_audio.size)
        X = np.append(X, new_audio);
        
print('Total Training datasize:',X.size/480,' -frames')
```

```python
Y=[]

for dirname, _, filenames in os.walk('txtfile/'):
    for filename in filenames:
        #print(os.path.join(dirname, filename))
        data= pd.read_csv(os.path.join(dirname, filename))
        temp = np.array(data.frequency)
        print(os.path.join(dirname, filename), 'Frames:', temp.size)
        Y = np.append(Y,temp)

minY = np.amin(Y)
for i in range(Y.size):
    Y[i] = Y[i] - minY
maxY = np.amax(Y)
print(minY, maxY)
X = np.reshape(X, (Y.size-1,480,1))

# file1 = open("MyFile.txt","a")
# for i in range(Y.size):
#     file1.write(np.array2string(Y[i]))
# file1.close()


print('Total labels:', Y.shape)
print('Total frames:', X.shape)
```

```python
#DFT of the framed data

# print(dim)
# XDFT = np.zeros([dim,1024,1])

# for i in range(dim):
# #     temp = np.fft.fft(X[i])
#     for j in range(480):
#         XDFT[i][j][0] = abs(temp[0][j]);

dim = X.shape[0]
new_data = np.zeros([dim, 1024, 1, 3])



for i in range(dim):
    for j in range(480):
        new_data[i][j][0][0] = X[i][j][0]

        
print(new_data.shape)
```

```python
new_data = np.reshape(new_data,[dim, 32, 32, 3])
    
print('Dataset shape:', new_data.shape)
print('Dataset newdata[0][0][0][0]:', new_data[0][0][0][0])
# data_trans = np.zeros([dim,32,32,3])
# for i in range(dim):
#     for j in range(32):
#         for k in range(32):
#             data_trans[i][j][k][0] = [i][j][k]

# print(XDFT_trans.shape)



Y = Y[0:Y.size-1]
print('Label data shape:', Y.shape)
```

```python
#split the training and testing dataset
from sklearn.model_selection import train_test_split

X_new = new_data[0:40000]
Y_new = Y[0:40000]

x_train,x_test,y_train,y_test=train_test_split(X_new,Y_new,test_size=0.2)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from keras.models import Sequential

from keras.layers import Dense, Flatten, Dropout

from keras.applications.vgg19 import VGG19

model = Sequential()
model.add(VGG19(weights="imagenet", include_top=False, input_shape=(32, 32, 3)))
model.add(Flatten())
model.add(Dense(1024, activation="relu"))
model.add(Dense(1024, activation="relu"))
model.add(Dense(1024, activation="relu"))
model.add(Dense(1952, activation="softmax"))

model.layers[-6].trainable = False

model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
train = model.fit(x_train, y_train, epochs=30)

model.save('TLModel')
```

```python
print('Testing data')
val_loss, val_acc = model.evaluate(x_test,y_test)
```

```python
sr, audio = wavfile.read('test_dataset/test.wav')
audio = zeropadding(audio, SamplePerWindow)

dim = int(audio.size/480)

test_audio = np.reshape(audio, (dim, 480, 1))
        
new_audio = np.zeros([dim, 1024, 1, 3])
for i in range(dim):
    for j in range(480):
        new_audio[i][j][0][0] = test_audio[i][j][0]
new_audio = np.reshape(new_audio, (dim, 32, 32, 3))

print(new_audio.shape)
test_output = model.predict(new_audio)
print(test_output.shape)
output = []
for i in range(dim):
    #print('max:', np.argmax(test_output[i][50:300]))
    result = np.where(test_output[i][50:500] == np.amax(test_output[i][80:500]))
    #print(result)
    output = np.append(output, result) 
    
file_TL = open("pitch_demo/TLModel_test.csv","w")
for i in range(dim):
    file_TL.write(str(output[i]))
    file_TL.write('\n')
file_TL.close()
print('Pitch demo file done')
```
