```python
from scipy.io import wavfile
#from os.path import expanduser
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def zeropadding(array, SamplePerWindow):
    if(array.size%SamplePerWindow != 0):
        left = array.size%SamplePerWindow
        add = np.zeros(SamplePerWindow - left)
        new_arr = np.append(array, add)
        return new_arr
    else:
        return array
```

```python
import python_speech_features

sr, audio = wavfile.read('new_dataset/mic_F01_si453.wav')
audio = zeropadding(audio, 480)

frames = int(audio.size/480)

mfcc_speech = python_speech_features.mfcc(signal=audio, samplerate=sr, winlen=0.01, winstep=0.005,nfft=1024)
print(mfcc_speech.shape)
temp = np.zeros((1,13))
mfcc_speech = np.append(mfcc_speech, temp)
print(mfcc_speech.size)
mfcc_speech = np.reshape(mfcc_speech,(frames,26))
print(mfcc_speech.shape)

audio = audio.reshape(frames, 480)
print(audio.shape)

X = []

for i in range(frames):
    feature_audio = np.append(audio[i], mfcc_speech[i])
    X.append(feature_audio)
    
X = np.array(X)
print(X.shape)
# X = []
# X = np.array(X)
# newdata = []
# i=0
# while i<mfcc_speech.shape[0]:
#     row = []
#     if i == mfcc_speech.shape[0]-1:
#         row.append(mfcc_speech[i])
#         row = np.append(row, mfcc_speech[i])
#     else:
#         sub = mfcc_speech[i] - mfcc_speech[i+1]
#         row.append(mfcc_speech[i])
#         row = np.append(row, sub)
#         newdata.append(row)
#         i = i+2
#     newdata = np.array(newdata)
            
#      X = np.append(X, newdata)
        
# X = np.array(X)
        
# print(X.shape)
```

```python
import os

X = []
temp = np.zeros((1,13))
SamplePerWindow = 480 ## sample rate 48000/s 0.01sec/window
for dirname, _, filenames in os.walk('new_dataset/'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
        sr, audio = wavfile.read(os.path.join(dirname,filename))
        new_audio = zeropadding(audio, SamplePerWindow)
        
        frames = int(new_audio.size/480)

        mfcc_speech = python_speech_features.mfcc(signal=new_audio, samplerate=sr, winlen=0.01, winstep=0.005,nfft=1024)
        mfcc_speech = np.append(mfcc_speech, temp)
        mfcc_speech = np.reshape(mfcc_speech, (frames, 26))
        
        new_audio = np.reshape(new_audio, (frames, 480))
        
        newdata = []
        for i in range(frames):
            feature_data = np.append(new_audio[i], mfcc_speech[i])
            X.append(feature_data)
            if(feature_data.size!=506):
                print('error shape not match')
        
X = np.array(X)
print('Total Training datasize:',X.shape,' -frames')
        
```

```python
Y=[]

for dirname, _, filenames in os.walk('txtfile/'):  ##Read the Y labels for the data
    for filename in filenames:
        #print(os.path.join(dirname, filename))
        data= pd.read_csv(os.path.join(dirname, filename))
        temp = np.array(data.frequency)
#         print(os.path.join(dirname, filename), 'Frames:', temp.size)
        Y = np.append(Y,temp)
print('Label file Y loaded')

minY = np.amin(Y)
for i in range(Y.size):
    Y[i] = Y[i] - minY
maxY = np.amax(Y)
print(minY, maxY)
Y = Y[0:Y.size-1]

print('Total labels:', Y.shape)
print('Total frames:', X.shape)
```

```python
from sklearn.model_selection import train_test_split


X_new = X[40000:160000]
Y_new = Y[40000:160000]



x_train,x_test,y_train,y_test=train_test_split(X_new,Y_new,test_size=0.1)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
SamplePerFrame = 480
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(512, activation=tf.nn.sigmoid)) 
model.add(tf.keras.layers.Dense(1024, activation=tf.nn.sigmoid)) 
model.add(tf.keras.layers.Dense(2048, activation=tf.nn.sigmoid))
model.add(tf.keras.layers.Dense(2048, activation=tf.nn.sigmoid)) 
model.add(tf.keras.layers.Dense(1952, activation=tf.nn.softmax)) 

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.build((1,506))

model.summary()
```

```python
model.fit(x_train, y_train, epochs=5) #15

model.save('MyModel_Feature')
```

```python
print('Testing data')
val_loss, val_acc = model.evaluate(x_test,y_test)
```
